Displaying IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
movie_data = requests.get(url).text
```

```python
movie_soup = BeautifulSoup(movie_data, "html.parser")
movie_ytags = {"class":"secondaryInfo"}
```

```python
movie_year = [tag.text for tag in movie_soup.find_all("span", movie_ytags)]
movie_year = [tag.replace("(", "").replace(")", "") for tag in movie_year]
del movie_year[100:]
movie_year
```

```python
release_year = pd.DataFrame(movie_year, index=range(1,101))
release_year.rename(columns= {0: "Release Year"}, inplace = True)
release_year
```

```python
movie_rtags = {"class":"ratingColumn imdbRating"}
movie_ratings = [tags.text for tags in movie_soup.find_all("td", movie_rtags)]
movie_ratings = [tags.replace("\n", "") for tags in movie_ratings]
del movie_ratings[100:]
movie_ratings
```

```python
imdb_ratings = pd.DataFrame(movie_ratings, index = range(1,101),columns=["Ratings"])
imdb_ratings
```

```python
imdb_100 = pd.merge(release_year, imdb_ratings, left_index=True, right_index=True)
imdb_100
```

```python
movie_name = [tags.text for tags in movie_soup.select(".titleColumn > a")]
del movie_name[100:]
imdb_name = pd.DataFrame(movie_name, index= range(1,101), columns= ["Title"])
imdb_name
```

```python
imdb_100 = pd.merge(imdb_name, imdb_100, left_index=True, right_index= True)
imdb_100.rename_axis("Ranking", inplace=True)
imdb_100
```
