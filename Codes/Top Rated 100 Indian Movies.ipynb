Displaying IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release)

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url  = "https://www.imdb.com/india/top-rated-indian-movies/"
html_data = requests.get(url).text
```

```python
indian_soup  = BeautifulSoup(html_data, "html.parser")
indian_rtag = {"class":"ratingColumn imdbRating"}
```

```python
indian_ratings = [tags.text for tags in indian_soup.find_all("td", indian_rtag)]
indian_ratings = [tags.strip() for tags in indian_ratings]
del indian_ratings[100:]
indian_ratings
```

```python
indian_name = [tags.text for tags in indian_soup.select(".titleColumn >a")]
del indian_name[100:]
indian_name
```

```python
indian_year = [tags.text for tags in indian_soup.select(".secondaryInfo")]
indian_year = [tags.replace("(", "").replace(")", "") for tags in indian_year]
del indian_year[100:]
indian_year
```

```python
indian_imdb = pd.DataFrame(list(zip(indian_name, indian_year, indian_ratings)), index = range(1,101), columns=["Title", "Release Year", "IMDb Rating"])
indian_imdb.rename_axis("Rank", inplace = True)
indian_imdb["IMDb Rating"] = indian_imdb["IMDb Rating"].astype(float)
indian_imdb["Release Year"] = indian_imdb["Release Year"].astype(int)
indian_imdb
indian_imdb.to_excel("Top Rated Indian Movies.xlsx")
```
