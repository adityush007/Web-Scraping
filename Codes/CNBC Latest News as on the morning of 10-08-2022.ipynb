Displaying Headline, Time and News link from <a>https://www.cnbc.com/world/?region=world</a>

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.cnbc.com/world/?region=world"
html_data = requests.get(url).text
news_soup = BeautifulSoup(html_data, "html.parser")
```

```python
news_time = [tags.text for tags in news_soup.select(".LatestNews-timestamp")]
news_time
```

```python
news_head = [tags.text for tags in news_soup.select(".LatestNews-headline")]
news_head
```

```python
news_table = news_soup.find("ul", class_="LatestNews-list")
news_link = [tags["href"] for tags in news_table.select("a[href]")]
news_link
```

```python
news_link = pd.DataFrame(news_link, columns=["Link"])
news_link
```

```python
news_link.drop([6,8,10,19,29,32], axis=0, inplace=True)
news_link.reset_index(drop=True, inplace=True)
news_link
```

```python
news_link.index = (range(1,31))
news_link
```

```python
news = pd.DataFrame(list(zip(news_head, news_time)), columns=["Headline", "Time"], index=range(1,31))
news
```

```python
news = pd.merge(news, news_link, left_index=True, right_index=True, )
news
```

```python
news.to_excel("CNBC Latest News.xlsx")
```

The code will require changes whenever we run it from top to bottom as the page is dynamic and latest news keeps on changing.
